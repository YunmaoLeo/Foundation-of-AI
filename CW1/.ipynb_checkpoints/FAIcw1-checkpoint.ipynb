### 8. Use  `report` and 5 different initial states, evluate heuristic function `h1` and `h2` of `EightPuzzle`, which one is better? provide your interpretation. (hint: you can define a new A* search function with a specified heuristic and pass the new search function to `report` function )  [3 Marks] 


```python
#A* search function using h2
def astar_tree_search_h2(problem):
    def h2(problem, node):
        """The Manhattan heuristic."""
        X = (0, 1, 2, 0, 1, 2, 0, 1, 2) #x coordinate of goal state 0,1,2,3,4,5,6,7,8
        Y = (0, 0, 0, 1, 1, 1, 2, 2, 2) #y coordinate of goal state 0,1,2,3,4,5,6,7,8
        return sum(abs(X[s] - X[g]) + abs(Y[s] - Y[g])
                   for (s, g) in zip(node.state, problem.goal) if s != 0)
    
    return best_first_tree_search(problem, f=lambda n: g(n) + h2(problem,n))
```

```python
show_stat_2 = report([astar_tree_search,astar_tree_search_h2],[e1,e2,e3,e4,e5])
show_bar(show_stat_2)
```

Answer: Using **h2** is better

The report shows that astar_tree_search_h2 genereate much less nods and goals than astar_tree_search using h1 with a largely shorter running time.

**Interpretation**: h2 measures the **manhattan distance** of inital nodes to their goals while h1 measure the **hamming distance**. In Eight Puzzle problem, using manhattan distance is more helpful to judge the difference between current state and the final state, thus astar_tree_search_h2 using manhattan distance has higher efficiency.
